```python

np.random.seed(42)

sales = np.random.randint(100, 1000, size=(5, 4))

sales[0]          # First product sales
sales[:, 1]       # Q2 sales for all products
sales[1:4, 2:4]   # Subset

total_sales = np.sum(sales)
product_wise = np.sum(sales, axis=1)
quarter_wise = np.sum(sales, axis=0)

np.mean(sales)
np.max(sales)
np.min(sales)
np.std(sales)

reshaped = sales.reshape(10, 2)
flattened = sales.flatten()

tax = np.array([10, 20, 30, 40])
final_sales = sales + tax

np.save("sales_data.npy", sales)
loaded_sales = np.load("sales_data.npy")

np.savez("data.npz", sales=sales, tax=tax)

```
